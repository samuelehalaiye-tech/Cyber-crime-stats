```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("all_recs.xlsx - Sheet1.csv", encoding='unicode_escape')
```

```python
df.head()
```

```python
df.tail(12)
```

```python
year_order= df['Year'].value_counts().index
#value count gets the unique occurnce of the individual years as well as arrnages them in order of frequency
location_order = df['Location'].value_counts().index
motive_order = df['Motive'].value_counts().index
industry_order = df['Industry'].value_counts().index
```

```python
plt.figure(figsize=(20,8))
plt.title("Attack Frequency by Year")
sb.countplot(data=df, y= 'Year', color=sb.color_palette()[2], order= year_order)
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(20,8))
plt.title("Attack Frequency by Industry")
sb.countplot(data=df, y= 'Industry', color=sb.color_palette()[2], order= industry_order)
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(20,8))
plt.title("Attack Frequency by Motive")
sb.countplot(data=df, y= 'Motive', color=sb.color_palette()[2], order= motive_order)
plt.xticks(rotation=45)
```
